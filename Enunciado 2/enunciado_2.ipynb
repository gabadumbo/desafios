**Enunciado 5:** Dado el archivo de registros de jugadas (eventos) de MemPy, mostrar un gr치fico con los usuarios que m치s veces jugaron el juego (independientemente del resultado de la partida.

Importo las librerias que voy a usar, pandas para la manipulaci칩n del dataset y matplotlib para el gr치fico del dataframe.

```python
import pandas as pd
from matplotlib import pyplot as plt
```

Abro el dataset con pandas

```python
arch = pd.read_csv('datos de prueba.csv')
```

Filtra por partidas, sacando los numeros de partida duplicados

```python
partidas = arch[arch['Partida'].duplicated() == False]
```

Luego uso value_counts() para ver la cantidad de veces que aparece cada nick en las diferentes partidas

```python
usuarios = partidas['Usuarie - nick'].value_counts()
```

Finalmente plotea el dataframe obtenido anteriormente, le agrega un titulo y algunas opciones esteticas.
La leyenda tapa parte del contenido al no estar en una ventana, se puede modificar y ver como queda con leyenda quitando el #

```python
usuarios.plot(kind='pie',autopct='%1.1f%%',shadow=True)
plt.title('Jugadores con mas partidas')
#plt.legend()
plt.axis('equal')
plt.show()
```

**Enunciado adicional:** Dado alguno de los datasets elegidos para el trabajo integrador, mostrar una nube de palabras (con la librer칤a wordcloud) con las palabras elegidas para el juego de acuerdo a alguno de los criterios elegidos en su trabajo.  Considerar utilizar la opci칩n de que se puedan repetir las palabras.


Importo las librerias que voy a usar

```python
import pandas as pd
from matplotlib import pyplot as plt
from wordcloud import WordCloud
```

Abro el archivo con pandas

```python
arch = pd.read_csv('lib.csv',encoding='utf-8')
```

Filtro el dataset por a침o 2018 y region LATAM

```python
paises = arch[arch.year == 2018]
region = paises[paises.region == 'Latin America & the Caribbean']
```

Luego ordeno el dataset por el puntaje en libertad que tengan los paises, y los paso a una lista

```python
ordenada = region.sort_values(by=['hf_score'],ascending=False)

l_nombres = ordenada.countries.to_list()
```

Luego los nombres de la lista se pasan a un solo string y se inicializa el wordcloud con los nombres y ciertas modificaciones

```python
str_unica = (' ').join(l_nombres)

wordcloud = WordCloud(max_font_size=50,background_color='white').generate(str_unica)
```

Se prepara el grafico con matplotlib y se muestra

```python
plt.figure()
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
